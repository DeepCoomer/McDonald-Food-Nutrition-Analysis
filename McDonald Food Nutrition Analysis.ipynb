```python
categories_by_saturated_fat = df[['Category', 'Saturated Fat']].groupby('Category').sum().sort_values('Saturated Fat', ascending=True).head(70)
```

```python
categories_by_saturated_fat.plot(kind='barh', figsize=(11,6))

plt.ylabel('Saturated Fat [grams]')
plt.title('Saturated Fat [grams] by Category');
```

```python
categories_by_sodium = df[['Category', 'Sodium']].groupby('Category').sum().sort_values('Sodium', ascending=True)
```

```python
categories_by_sodium.plot(kind='bar', figsize=(11,6))

plt.ylabel('Sodium [mg]')
plt.title('Sodium [mg] by Category');
```

```python
items_by_sodium = df[['Item', 'Sodium']].groupby('Item').sum().sort_values(['Sodium', 'Item'], ascending=True).head(20)
items_by_sodium.plot(kind='bar', figsize=(21,7))

plt.ylabel('Sodium [in mg]')
plt.title('Sodium [in mg] by Item');
```

```python
categories_by_sugars = df[ ['Category', 'Sugars'] ].groupby('Category').sum().sort_values('Sugars', ascending=True).head(70)

categories_by_sugars.plot(kind='bar', figsize=(11,6))

plt.ylabel('Sugars [grams]')
plt.title('Sugars [grams] by Category')
```

```python
items_by_sugars = df[ ['Item', 'Sugars'] ].groupby('Item').sum().sort_values(['Sugars', 'Item'], ascending=True).head(80)

items_by_sugars.plot(kind='bar', figsize=(21,7))

plt.ylabel('Sugars [in grams]')
plt.title('Sugars [in grams] by Item');
```

```python
categories_by_dietary_fiber = df[ ['Category', 'Dietary Fiber'] ].groupby('Category').sum().sort_values('Dietary Fiber', ascending=True).head(70)


categories_by_dietary_fiber.plot(kind='pie', subplots=True, figsize=(11,6))

plt.ylabel('Dietary Fiber [[in grams]')
plt.title('Dietary Fiber [[in grams] by Category');
```
